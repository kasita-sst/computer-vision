# Image Filtering

Changes the range of an image (pixel values), not changing any of the position

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
import cv2
import matplotlib.pyplot as plt

def rgbplot(image):
    # Convert color spaces
    convert_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    # Render the image data
    plt.imshow(convert_image)
    # Display the plot window
    plt.show()
```

## Linear Image Filter

### Mean Filter / Mean Pooling<br>
- the larger the kernel_size are, the more blur our image looks

```python
def mean_filter(image, kernel_size, stride_num=1):
    padded_image = cv2.copyMakeBorder(image,
                                      kernel_size//2,
                                      kernel_size//2,
                                      kernel_size//2,
                                      kernel_size//2,
                                      cv2.BORDER_CONSTANT)
    
    filtered_image = np.zeros_like(image)

    #its a 2D image, so nested for loop iterates over every pixel
    #stride default is 1, its a step parameter
    for i in range(0, image.shape[0], stride_num):
        for j in range(0, image.shape[1], stride_num):
            #finding mean value for each filter
            window = padded_image[i:i+kernel_size, j:j+kernel_size]
            mean_val = np.mean(window)
            filtered_image[i, j] = mean_val

    return filtered_image

image = cv2.imread('../img/anxious-cell.png', cv2.IMREAD_GRAYSCALE)

filtered_image = mean_filter(image, 10, 1)

rgbplot(filtered_image)
```

```python
filtered_image_kernel50 = mean_filter(image, 50, 1)

rgbplot(filtered_image_kernel50)
```

```python
filtered_image_kernel10_stride5 = mean_filter(image, 50, 2)

rgbplot(filtered_image_kernel10_stride5)
```
