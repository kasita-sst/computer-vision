```python
import cv2
import matplotlib.pyplot as plt
```

# Display Image

### Reading GrayScale

```python
image_grayscale = cv2.imread('../img/anxious-cell.png', cv2.IMREAD_GRAYSCALE)
```

```python
image_convert= cv2.cvtColor(image_grayscale, cv2.COLOR_BGR2RGB)
plt.imshow(image_convert)

print(image_convert.shape)
```

### Reading Color Image

```python
image_color = cv2.imread('../img/anxious-cell.png', cv2.IMREAD_COLOR)

plt.imshow(image_color)
print(image_color.shape)
```

```python
image_convert= cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)

plt.imshow(image_color)
print(image_color.shape)
```
